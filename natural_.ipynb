2. Model Implementation

```python
# Encode the Target Column (Alphabet Letters)

df['letter'] = df['letter'].astype('category').cat.codes

df.head()
```

```python
# Split Features (X) and Target (y)

X = df.drop('letter', axis=1)
y = df['letter']


```

```python
# Normalize the Data (0–1 Range)
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Train–Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)


```

```python
# Build the ANN Model (With Hidden Layer)
model = Sequential()

# Hidden Layer (32 neurons)
model.add(Dense(32, activation='relu', input_shape=(X_train.shape[1],)))

# Output Layer (26 classes)
model.add(Dense(26, activation='softmax'))

```

```python
# Compile the Model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)


```

```python
# Train the Model
history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.2
)


```

```python
# Evaluate on Test Data
test_loss, test_acc = model.evaluate(X_test, y_test)
print("Test Accuracy:", test_acc)

```

3. Hyperparameter Tuning

```python
# QUICK MANUAL EXPERIMENTS (fast checks)

def build_model_simple(neurons=32, activation='relu', lr=0.001, hidden_layers=1):
    model = Sequential()
    model.add(Dense(neurons, activation=activation, input_shape=(X_train.shape[1],)))
    for _ in range(hidden_layers-1):
        model.add(Dense(neurons, activation=activation))
    model.add(Dense(26, activation='softmax'))
    model.compile(optimizer=Adam(learning_rate=lr),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

results = []
configs = [
    {'hidden_layers':1,'neurons':32,'lr':1e-3},
    {'hidden_layers':2,'neurons':32,'lr':1e-3},
    {'hidden_layers':1,'neurons':64,'lr':1e-3},
    {'hidden_layers':1,'neurons':32,'lr':1e-2},
]

for cfg in configs:
    model = build_model_simple(**cfg)
    history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2, verbose=0)
    loss, acc = model.evaluate(X_test, y_test, verbose=0)
    results.append({**cfg, 'test_acc':acc})
    print(cfg, "-> test_acc:", acc)

df_results = pd.DataFrame(results)
print(df_results)

```
